```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from IPython import display
import time
```

Define kernels and kernel gradient.

```python
def ker(x,y,alpha):
    # RBF exp(alpha * |x-y|^2)
    return np.exp(-alpha * np.sqrt((x - y)**2))

def ker_grad(x,y,alpha):
    # gradient of ker(x,y)=RBF exp(alpha * |x-y|^2)
    return -alpha * 2 * (x - y) * ker(x,y,alpha)
```

Now we define our posterior. We will just define p(x) as a gaussian distribution. 

The goal of SVGD is to transform points from a arbitrary distribution to fit the chosen p(x). However we don't have a direct way of sampling and computing p(x) we only have its gradients. 

SVGD:s goal is similar to what MCMC sampling aims to do. 

```python
def normal_dist(x,mu=0,std=1):
    return (1/(std * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x-mu)/std)**2)

def grad_normal_dist(x,mu=0,std=1):
    return -(1/std) * (x-mu) * (1/(std * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((x-mu)/std)**2)
```

Plot of p(x) and d/dx p(x)

```python
x_lin = np.linspace(-5, 5, num=50)
y_lin = [normal_dist(xi) for xi in x_lin]
plt.plot(x_lin, y_lin)
```

```python
y_lin_grad = [grad_normal_dist(xi) for xi in x_lin]
plt.plot(x_lin, y_lin_grad)
```

Now we define a gradient descent function that will transport along the gradients to a local mimimum of p(x).

```python
def forward(x,num_itr=50,alpha=1,step_sz=1):
    plt.plot(x_lin, y_lin)
    
    x_steps = []
    x_steps.append(x)
    
    for i in range(num_itr):
        x_ker = np.zeros((len(x),len(x)))
        x_ker_grad = np.zeros((len(x),len(x)))
        
        for p in range(x.shape[0]):
            x_ker[p] = ker(x[p],x,alpha)
            x_ker_grad = ker_grad(x[p],x,alpha)
        
        x_grad = grad_normal_dist(x)
        
        x = x + step_sz * x_grad# np.mean(x_grad * x_ker) + x_ker_grad
        
        y = normal_dist(x)
        plt.plot(x, y, 'r*')
        display.clear_output(wait=True)
        display.display(plt.gcf())
        time.sleep(0.01)
        
        plt.clf()
        plt.cla()
        plt.plot(x_lin, y_lin)
        
        x_steps.append(x)

    return x_steps
```

```python
x_lin_in = np.zeros(1)
x_lin_in[0] = -3 #np.linspace(-5, -5.1, num=1)
```

```python
x_res = forward(x_lin_in,num_itr=20,alpha=1,step_sz=5)
```

Seems to work!

This should spread out the particles as much as possible to inf. 

```python
def forward_svgd(x,num_itr=100,alpha=1,step_sz=1):
    x_steps = []
    x_steps.append(x)
    plt.plot(x_lin, y_lin)
    colors = cm.rainbow(np.linspace(0, 1, len(x)))

    for i in range(num_itr):
        x_ker = np.zeros((len(x),len(x)))
        x_ker_grad = np.zeros(len(x))
        
        for p in range(x.shape[0]):
            x_ker[p] = ker(x[p],x,alpha)
            x_ker_grad[p] = np.mean(ker_grad(x[p],x,alpha),-1)

        x_grad = grad_normal_dist(x)
        
        x = x + step_sz * (-1) * x_ker_grad#(np.mean(x_grad * x_ker,0) + )

        y = normal_dist(x)

        for i in range(len(colors)):            
            plt.plot(x[i], y[i], '*',color=colors[i])
        
        display.clear_output(wait=True)
        display.display(plt.gcf())

        if num_itr < 200:
            time.sleep(0.00001)
        
        plt.clf()
        plt.cla()
        plt.plot(x_lin, y_lin)
        
        x_steps.append(x)

    return x_steps
```

Lets now do SVGD with only the first term. This should pull the samples to the max. 

We add a orange histogram to track the distribution of the samples. 

```python
def forward_svgd(x,num_itr=100,alpha=1,step_sz=1):
    x_steps = []
    x_steps.append(x)
    plt.plot(x_lin, y_lin)
    colors = cm.rainbow(np.linspace(0, 1, len(x)))

    for i in range(num_itr):
        x_ker = np.zeros((len(x),len(x)))
        x_ker_grad = np.zeros(len(x))
        
        for p in range(x.shape[0]):
            x_ker[p] = ker(x[p],x,alpha)
            x_ker_grad[p] = np.mean(ker_grad(x[p],x,alpha),0)
        
        x_grad = grad_normal_dist(x)
        x = x + step_sz * (np.mean(x_grad * x_ker,0) ) #+ (-1) * x_ker_grad# )

        y = normal_dist(x)
        
        for i in range(len(colors)):            
            plt.plot(x[i], y[i], '*',color=colors[i])
        
        weights = (np.ones_like(x) / len(x))
        plt.hist(x, 25, weights=weights * 0.5)
            
        display.clear_output(wait=True)
        display.display(plt.gcf())

        if num_itr < 200:
            time.sleep(0.00001)
        
        plt.clf()
        plt.cla()
        plt.plot(x_lin, y_lin)
        
        x_steps.append(x)

    return x_steps
```

Now we do a stein gradient decent. Here we transport a number of particles to fit our 'unknown' distribution. 

Lets also start with more samples points.

```python
def forward_svgd(x,num_itr=100,alpha=1,step_sz=1):
    x_steps = []
    x_steps.append(x)
    plt.plot(x_lin, y_lin)
    colors = cm.rainbow(np.linspace(0, 1, len(x)))

    for i in range(num_itr):
        x_ker = np.zeros((len(x),len(x)))
        x_ker_grad = np.zeros(len(x))
        
        for p in range(x.shape[0]):
            x_ker[p] = ker(x[p],x,alpha)
            x_ker_grad[p] = np.mean(ker_grad(x[p],x,alpha),0)
        
        x_grad = grad_normal_dist(x)
        x = x + step_sz * (np.mean(x_grad * x_ker,0) - x_ker_grad)

        y = normal_dist(x)

        for i in range(len(colors)):            
            plt.plot(x[i], y[i], '*',color=colors[i])
        
        weights = (np.ones_like(x) / len(x))
        plt.hist(x, 25, weights=weights * 0.5)
            
        display.clear_output(wait=True)
        display.display(plt.gcf())

        #if num_itr < 200:
        #    time.sleep(0.00001)
        
        plt.clf()
        plt.cla()
        plt.plot(x_lin, y_lin)
        
        x_steps.append(x)

    return x_steps
```

```python
x_lin_in = np.linspace(-3, 3, num=200)
x_res = forward_svgd(x_lin_in,num_itr=100,alpha=0.1,step_sz=1)

print('Mean and std of samples (should be 0/1): ',x_res[-1].mean(), x_res[-1].std())
```
